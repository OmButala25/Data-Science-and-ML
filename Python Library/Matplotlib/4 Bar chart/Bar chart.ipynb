**Matplotlib Tutorial: Bar Chart**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Simple bar chart showing revenues of major US tech companies

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
#it wont honor string values inside any visualization we should first convert into array then chart it
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue, label="Revenue") #here we have 0 to 4 values of x-axis not the company name
```

```python
# Get or set the current tick locations and labels of the x-axis.
# here we will convert the numbers into strind using xticks

plt.xticks(xpos, company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')

plt.bar(xpos,revenue, label="Revenue")
plt.legend()
```

**Multiple Bars** showing revenue and profit of major US tech companies

```python
profit=[40,2,34,12]
```

```python
# plt.bar(xpos-0.2,revenue, label="Revenue")
# plt.bar(xpos+0.2,profit, label="Profit")

plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

Horizontal bar chart using **barh** function

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

```python
plt.barh(xpos-0.2,revenue, height=0.4, label="Revenue")
plt.barh(xpos+0.2,profit, height=0.4, label="Profit")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```
