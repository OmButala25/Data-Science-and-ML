**Matplotlib Tutorial: Pie Charts**

Pie charts show the size of items (called wedge) in one data series, proportional to the sum of the items. The data points in a pie chart are shown as a percentage of the whole pie. Matplotlib API has a pie() function that generates a pie diagram representing data in an array.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Draw pie chart to track down home expenses

exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels = exp_labels)
plt.show()
```

```python
# Draw pie chart as perfect circle
plt.pie(exp_vals, labels = exp_labels, shadow= True)
plt.axis("equal")
plt.show()
```

```python
# Show percentages for every pie. Specify radius to increase chart size

plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, autopct='%0.2f%%', radius=2) #autopct to get the percent value of each part and thats the format %f%%
plt.show()
```

```python
# Explode to cut out 
plt.axis("equal")
plt.pie(exp_vals, labels= exp_labels, autopct='%0.1f%%', explode=[0,0.2,0,0,0.1]) #explode to get 0.2 and 0.1 out of the pie
plt.show()
```

```python
# Counterclock and angle properties

plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45) #startangle to change the angle or to change 
plt.show()
```
