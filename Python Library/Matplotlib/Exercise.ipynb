Using following static data for companies revenues,

company=['GOOGL','AMZN','MSFT','FB']

revenue=[90,136,89,27]

Plot a bar chart that looks like below

Save this chart as jpg file and then pdf file with 5 inch of padding

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.xlabel('Company')
plt.ylabel('Revenue(Bln)')
plt.title('US Technology Stocks')
plt.bar(company, revenue, label='revenue')
plt.legend(fontsize = 15)

plt.savefig("US stocks.jpg", bbox_inches= 'tight', pad_inches = 5, transparent = True)
```

```python
plt.xlabel('Company')
plt.ylabel('Revenue(Bln)')
plt.title('US Technology Stocks')
plt.bar(company, revenue, label='revenue')
plt.legend(fontsize = 15)

plt.savefig("US stocks.pdf", bbox_inches= 'tight', pad_inches = 5, transparent = True)
```
