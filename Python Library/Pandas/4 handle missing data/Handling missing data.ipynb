```python
import pandas as pd
df = pd.read_csv('weather_data.csv', parse_dates = ['day']) #parse to convert day column which was in str to date format
type(df.day[0]) #checked the type
df
```

```python
df.set_index('day',inplace=True) #setted index as day and modified the original dataset using inplace=true
df
```

FillNa() funtion to fill NaN/ Fill all NaN with one specific value

```python
new_df = df.fillna(0)
new_df
```

Fill na using column names and dict

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'No Event'
})
new_df
```

Use method to determine how to fill na values/ to carry forward the estimates 

```python
# new_df = df.fillna(method="ffill")
new_df = df.ffill() #forward fill
# new_df = df.bfill() #backward fill
new_df
```

Use of axis

```python
# new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df = df.bfill(axis = "columns") #new version method
new_df
```

limit parameter

```python
new_df = df.ffill(limit=1) #if their is gap more than 1 and we set limit as 1 it will fill only once next fill and then NaN
new_df
```

Interpolation - Interpolation is one such method of filling data. Interpolation is a technique in Python used to estimate unknown data points between two known data points. Interpolation is mostly used to impute missing values in the dataframe or series while pre-processing data.
helps to determine the data points in between the given data ones.

```python
import warnings
warnings.filterwarnings('ignore')

new_df = df.infer_objects(copy=False)
new_df = df.interpolate(method = 'linear')
new_df
```

```python
new_df = df.interpolate(method="time") 
new_df
#Notice that in above temperature on 2017-01-04 is 29 instead of 30 (in plain linear interpolate)
#There are many other methods for interpolation such as quadratic, piecewise_polynomial, cubic etc. Just google "dataframe interpolate" to see complete documentation
```

Dropna() - The dropna() method removes the rows that contains NULL values. The dropna() method returns a new DataFrame object unless the inplace parameter is set to True , in that case the dropna() method does the removing in the original DataFrame instead.

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
#if i want to keep NaN having in single col in whole row then we use
new_df = df.dropna(thresh=1)
new_df
```

```python
# here we will use some replacing methods one by one
import numpy as np
df1 = pd.read_csv("weather.csv")
df1
```

Replacing single value method

```python
new_df1 = df1.replace(-99999, value=np.NaN)
new_df1
```

Replacing list with single value

```python
new_df1 = df1.replace(to_replace=[-99999,-88888], value=0)
new_df1
```

Replacing per column

```python
new_df1 = df1.replace({
        'temperature': -99999,
        'windspeed': -99999,
        'event': '0'
    }, np.nan)
new_df1
```

Replacing by using mapping

```python
new_df1 = df1.replace({
        -99999: np.nan,
        'no event': 'Sunny',
    })
new_df1
```

Regex - A Regular Expression (or Regex) is a pattern (or filter) that describes a set of strings that matches the pattern. In other words, a regex accepts a certain set of strings and rejects the rest.

```python
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
# remove mph and F char value we gonna remove
new_df1 = df1.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True) 
new_df1
```

Replacing list with another list

```python
df1 = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df1
```

```python
new_df1 = df1.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4])
new_df1
```
